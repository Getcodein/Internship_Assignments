```python
##------------answer 10------------------------------------
import selenium
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import ElementNotInteractableException,NoSuchElementException
from selenium.webdriver.common.by import By
import pandas as pd
import time


#connecting to webdriver

driver=webdriver.Chrome(r"chromedriver.exe")


# clicking the search button
try:
    driver.get("https://www.hostelworld.com/")
    # finding element for laptop search bar
    srch_jb = driver.find_element(By.XPATH,'//input[@class="location-text"]')
    srch_jb.send_keys("London")
    time.sleep(1)
    
    search_btn=driver.find_element(By.XPATH,'/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/button')
    search_btn.click()
    
    search_button1=driver.find_element(By.XPATH,'//ul[@id="predicted-search-results"]/li')
    search_button1.click()

except NoSuchElementException as e:
    driver.get("https://www.hostelworld.com/s?q=London,%20England&country=England&city=London&type=city&id=3&from=2022-12-20&to=2022-12-23&guests=2&page=1")

time.sleep(7)

# creating empty lists for scraping data
h_name=[]
distance=[]
rating=[]
t_review=[]
o_review=[]
private_price=[]
dorm_price=[]
facilities=[]
pr_desc=[]

h_tags=driver.find_elements(By.XPATH,'//div[@class="property"]/div[1]/h2/a[1]')
distance_tags=driver.find_elements(By.XPATH,'//div[@class="property"]/div/div[1]/a/span[1]')
rating_tags=driver.find_elements(By.XPATH,'//div[@class="bottom-rating"]/a/div[1]/div[1]')
t_tags=driver.find_elements(By.XPATH,'//div[@class="bottom-rating"]/a/div[1]/div[2]/div[2]')
o_tags=driver.find_elements(By.XPATH,'//div[@class="bottom-rating"]/a/div[1]/div[2]/div/span')
private_tags=driver.find_elements(By.XPATH,'//div[@class="prices-col"]/a/div[1]')
dorm_tags=driver.find_elements(By.XPATH,'//div[@class="prices-col"]/a/div[2]')
facilities_tags=driver.find_elements(By.XPATH,'//div[@class="facilities-label facilities"]')
Pr_tags=driver.find_elements(By.XPATH,'//div[@class="rating-factors prop-card-tablet rating-factors small"]')


print(len(h_tags),len(distance_tags),len(rating_tags),len(t_tags),len(o_tags),len(private_tags),len(dorm_tags),len(facilities_tags),len(Pr_tags))

for k in h_tags:
    m=k.text
    h_name.append(m)

for k1 in distance_tags:
    m1=k1.text
    distance.append(m1)
    

for k2 in rating_tags:
    m2=k2.text
    rating.append(m2)
    

for k3 in t_tags:
    m3=k3.text
    t_review.append(m3)

for k4 in o_tags:
    m4=k4.text
    o_review.append(m4)

for k5 in private_tags:
    m5=k5.text.replace('\n','|')
    private_price.append(m5)
    
for k6 in dorm_tags:
    m6=k6.text.replace('\n','|')
    dorm_price.append(m6)
    
for k7 in facilities_tags:
    m7=k7.text.replace('\n','|')
    facilities.append(m7)
    
for k8 in Pr_tags:
    m8=k8.text.replace('\n','|')
    pr_desc.append(m8)

print(len(h_name),len(distance),len(rating),len(t_review),len(o_review),len(private_price),len(dorm_price),len(facilities),len(pr_desc))
```

```python
#-------------------creating data frame -------------------------
shop=pd.DataFrame({})
shop['hostel_name']=h_name
shop['distance']=distance
shop['ratings']=rating
shop['totalreviews']=t_review
shop['overallreviews']=o_review
shop['privatesfrom_price']=private_price
shop['dormsForm_price']=dorm_price
shop['facilities']=facilities
shop['property_des']=pr_desc



shop
```
